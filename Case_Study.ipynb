### Imporing Libraries

```python
import pandas as pd
from matplotlib import pyplot as plt
```

### Loading Data

```python
df=pd.read_csv('Salaries.csv')
df
```

```python
df.info()
```

### 1.	How much the total salary has increased from year 2011 to 2014?

```python
total_salary_increase = df[df['Year'] == 2014]['TotalPayBenefits'].sum() - df[df['Year'] == 2011]['TotalPayBenefits'].sum()
```

```python
print('Total salary increase: ', total_salary_increase)
```

### 2.	Which job title had the highest mean salary in the years 2011 through 2014? 

```python
highest_mean_salary_job = df.groupby('JobTitle')['TotalPayBenefits'].mean().idxmax()
```

```python
print('Highest mean salary job: ', highest_mean_salary_job)
```

### 3.	How much money could have been saved in the year 2014 by stopping overtime pay?

```python
overtime_saved_2014 = df[df['Year'] == 2014]['OvertimePay'].sum()
```

```python
print('Overtime saved 2014: ', overtime_saved_2014)
```

### 4.	Who were the top earning employees across all the years? 

```python
top_earning_employees = df.sort_values(by= 'TotalPayBenefits', ascending = False).head(3)['EmployeeName'].tolist()
```

```python
print('Top earning employees: ', top_earning_employees)
```

### 5.	Which were the least 5 common jobs in 2014, and how much did they cost? 

```python
least_common_jobs_2014_ = df[df['Year'] == 2014]['JobTitle'].value_counts().sort_values().head(5)
```

```python
print('Least 5 common jobs in 2014:\n', least_common_jobs_2014_)
```

```python
df[df['Year'] == 2014]['JobTitle'].value_counts()
```

```python
least_common_jobs_2014 = df[df['Year'] == 2014]['JobTitle'].value_counts()
```

```python
print('Least 5 common jobs in 2014:\n', least_common_jobs_2014)
```

```python
least_common_jobs_cost = df[df['JobTitle'].isin(least_common_jobs_2014)].groupby('JobTitle')['TotalPayBenefits'].sum()
```

```python
print('Least common jobs cost:', least_common_jobs_cost)
```

### 6.	What percentage did overtime pay constitute of the total pay? 

```python
overtime_percentage = (df['OvertimePay'].sum() / df['TotalPayBenefits'].sum()) * 100
```

```python
print("Overtime percentage: {:.2f}%".format(overtime_percentage))
```

### Data Visualization

1. Plot the distribution of Base Pay

```python
plt.figure(figsize=(10,6))
plt.hist(df['BasePay'], bins= 30, color= 'blue', alpha= 0.6, edgecolor= 'black')
plt.xlabel('Base Pay')
plt.ylabel('Frequency')
plt.title('Distribution of Base Pay')
plt.show()
```

The most common base pay between 0 and 100000

2. Plot the average base pay over years

```python
plt.figure(figsize=(10,6))
yearly_data = df.groupby('Year')['BasePay'].mean()
plt.plot(yearly_data.index, yearly_data.values, marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Average Base Pay (Dollars)')
plt.title('Average Base Pay Over Years')
plt.xticks(yearly_data.index)
plt.show()
```

Year 2013 shows the highest average base pay

3. Plot the total pay benefits over years

```python
plt.figure(figsize=(10,6))
yearly_data_ = df.groupby('Year')['TotalPayBenefits'].sum()
plt.plot(yearly_data_.index, yearly_data_.values, marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Total Pay Benefits (Dollars)')
plt.title('Total Pay Benefits Over Years')
plt.xticks(yearly_data_.index)
plt.show()
```

Between 2011 and 2012 there was a significant increase in total pay benefits 

4. Plot the total overtime pay over years

```python
plt.figure(figsize=(10,6))
yearly_data0 = df.groupby('Year')['OvertimePay'].sum()
plt.plot(yearly_data0.index, yearly_data0.values, marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Total Overtime Pay (Dollars)')
plt.title('Total Overtime Pay Over Years')
plt.xticks(yearly_data0.index)
plt.show()
```

The total overtime pay increases over the years

5. Plot the top 10 common jop titles by average base pay

```python
plt.figure(figsize=(12,8))
job_title_data = df.groupby('JobTitle')['BasePay'].mean().sort_values(ascending = False).head(10)
job_title_data.plot(kind ='barh', color='blue')
plt.xlabel('Average Base Pay (Dollars)')
plt.ylabel('Job Title')
plt.title('Top 10 Job Titles by Average Base Pay')
plt.gca().invert_yaxis()
plt.show()
```

The most common job title by average base pay is Chief of Police


6. Plot the top 10 employee name by average base pay

```python
plt.figure(figsize=(10,6))
employee_name_data = df.groupby('EmployeeName')['BasePay'].mean().sort_values(ascending = False).head(10)
plt.barh(employee_name_data.index, employee_name_data.values, color='blue')
plt.xlabel('Average Base Pay (Dollars)')
plt.ylabel('Employee Name')
plt.title('Top 10 Employee Name by Average Base Pay')
plt.gca().invert_yaxis()
plt.show()
```

The employee name with the highest average base pay is Gregory P Suhr

7. Plot the top 10 common jop titles

```python
plt.figure(figsize=(10,6))
most_common_jobs = df['JobTitle'].value_counts().head(10)
plt.barh(most_common_jobs.index, most_common_jobs.values, color='blue')
plt.xlabel('Count')
plt.ylabel('Job Title')
plt.title('Top 10 Common Job Titles')
plt.gca().invert_yaxis()
plt.show()
```

The most common job title is Transit Operator

8. Plot the total other pay over years

```python
plt.figure(figsize=(10,6))
yearly_data_1 = df.groupby('Year')['OtherPay'].sum()
plt.plot(yearly_data_1.index, yearly_data_1.values, marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Total Other Pay (Dollars)')
plt.title('Total Other Pay Over Years')
plt.xticks(yearly_data_1.index)
plt.show()
```

Year 2013 shows the highest total other pay
